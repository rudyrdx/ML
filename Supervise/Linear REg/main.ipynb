```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

## Dataloading and Cleaning

```python
data = pd.read_csv('Salary_dataset.csv')
data.info()
```

```python
data.dropna(inplace=True)
data.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
data.head()
```

```python
X = data[['YearsExperience']]
y = data['Salary']
```

## Setting the Model

```python
model = LinearRegression()
model.fit(X, y)
```

```python
years_range = np.linspace(10, 1, 1000).reshape(-1, 1)
# we used linspace to create an array of 1000 
# numbers between 10 and 1 because we want to 
# predict the salary for the next 10 years
```

```python
salary_pred = model.predict(years_range)
```

## Plotting the graph

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsExperience', y='Salary', data=data, color='blue', s=100)
plt.plot(years_range, salary_pred, color='red', label='Linear Regression')
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
```

## Prediction

```python
prediction = model.predict([[1]])
print(f"Predicted Salary for 1 years of experience: ${prediction[0]:.2f}")
```
