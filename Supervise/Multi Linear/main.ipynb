```python
import pandas as pd

df = pd.read_csv('50_Startups.csv')

df.info()
```

```python
df.describe()
```

```python
# R&D Spend	Administration	Marketing Spend	Profit
# correlation matrix
df.corr()
```

```python
#train test split
from sklearn.model_selection import train_test_split
# multiple linear regression
from sklearn.linear_model import LinearRegression
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(df[['R&D Spend', 'Administration', 'Marketing Spend']], df['Profit'], test_size=0.2, random_state=0)
```

```python
model = LinearRegression()

model.fit(Xtrain, ytrain)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='R&D Spend', y='Profit', data=df)
sns.scatterplot(x='Administration', y='Profit', data=df)
sns.scatterplot(x='Marketing Spend', y='Profit', data=df)
# legends
plt.legend(['R&D Spend', 'Administration', 'Marketing Spend'])
# regression line
plt.plot(df['R&D Spend'], model.predict(df[['R&D Spend', 'Administration', 'Marketing Spend']]), color='red')
plt.plot(df['Administration'], model.predict(df[['R&D Spend', 'Administration', 'Marketing Spend']]), color='green')
plt.plot(df['Marketing Spend'], model.predict(df[['R&D Spend', 'Administration', 'Marketing Spend']]), color='blue')
plt.show()
```
