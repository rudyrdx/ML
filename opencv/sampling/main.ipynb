# Sampling

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = 'bruce.jpg'
img = cv2.imread(image)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
print(img.shape)
```

```python
downsampled = cv2.pyrDown(img)
print(downsampled.shape)
```

```python
upsampled = cv2.pyrUp(img)
upsampled = cv2.pyrUp(upsampled)
upsampled = cv2.pyrUp(upsampled)
print(upsampled.shape)
```

```python
cv2.imwrite('upsampled.jpg', upsampled)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(img)
axs[0].set_title('Original Image')

axs[1].imshow(downsampled)
axs[1].set_title('Downsampled Image')

axs[2].imshow(upsampled)
axs[2].set_title('Upsampled Image')
plt.show()
```

# Quantization

```python
import numpy as np
```

```python
image = cv2.imread(image, cv2.IMREAD_GRAYSCALE)
```

```python
# Define the number of quantization levels
num_levels = 4

# Compute the quantization step size
step_size = 256 // num_levels

# Apply uniform quantization
quantized_image = (image // step_size) * step_size
```

```python
# Define the number of quantization levels
num_levels = 6

# Compute the quantization step size
step_size = 256 // num_levels

# Apply uniform quantization
quantized_image = (image // step_size) * step_size
```

```python
# Display the original and quantized images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f'Quantized Image with {num_levels} Levels')
plt.imshow(quantized_image, cmap='gray')
plt.axis('off')

plt.show()
```
