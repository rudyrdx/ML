### Image Transformations

```python
#linear transformation
#logarithimic transformation
# powerlaw transformation
# gamma transformation
# negative transformation
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
img = cv.imread('image.jpg', 0)
def linear_transformation(img, alpha, beta):
    new_img = alpha * img + beta
    new_img = np.clip(new_img, 0, 255)
    new_img = new_img.astype(np.uint8)
    return new_img

def logarithmic_transformation(img, c):
    new_img = c * np.log(1 + img)
    new_img = np.clip(new_img, 0, 255)
    new_img = new_img.astype(np.uint8)
    return new_img

def powerlaw_transformation(img, gamma):
    new_img = np.power(img, gamma)
    new_img = np.clip(new_img, 0, 255)
    new_img = new_img.astype(np.uint8)
    return new_img

def gamma_transformation(img, gamma):
    new_img = np.power(img, 1/gamma)
    new_img = np.clip(new_img, 0, 255)
    new_img = new_img.astype(np.uint8)
    return new_img

def negative_transformation(img):
    new_img = 255 - img
    return new_img


# Linear Transformation
linear_img = linear_transformation(img, alpha=1.5, beta=50)
plt.subplot(2, 3, 1)
plt.imshow(linear_img, cmap='gray')
plt.title('LinearT')

# Logarithmic Transformation
log_img = logarithmic_transformation(img, c=1.5)
plt.subplot(2, 3, 2)
plt.imshow(log_img, cmap='gray')
plt.title('LogT')

# Power Law Transformation
power_img = powerlaw_transformation(img, gamma=0.5)
plt.subplot(2, 3, 3)
plt.imshow(power_img, cmap='gray')
plt.title('PowerT')

# Gamma Transformation
gamma_img = gamma_transformation(img, gamma=1.5)
plt.subplot(2, 3, 4)
plt.imshow(gamma_img, cmap='gray')
plt.title('GammaT')

# Negative Transformation
negative_img = negative_transformation(img)
plt.subplot(2, 3, 5)
plt.imshow(negative_img, cmap='gray')
plt.title('NegativeT')

plt.tight_layout()
plt.show()


```
