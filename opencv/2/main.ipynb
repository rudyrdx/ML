```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

image = cv2.imread('image.jpg')
original_image = image
# Display the noisy image
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.axis('off')
plt.title("Noisy Image for Denoising Techniques")
plt.show()
```

```python
def degradation(image):
    # Apply degradation formula
    degraded_image = image + np.random.normal(0, 10, image.shape)
    return degraded_image

def restoration(degraded_image):
    # Apply restoration formula
    restored_image = degraded_image - np.random.normal(0, 10, degraded_image.shape)
    return restored_image
#
# Degrade the original image
degraded_image = degradation(original_image)

# Restore the degraded image
restored_image = restoration(degraded_image)

# Plot the images
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(original_image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(degraded_image, cmap='gray')
axes[1].set_title('Degraded Image')
axes[1].axis('off')

axes[2].imshow(restored_image, cmap='gray')
axes[2].set_title('Restored Image')
axes[2].axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt

import cv2
import numpy as np

image = cv2.imread("image.jpg")

sharpen_kernel = np.array([
    [0, -1, 0],
    [-1, 8, -1],
    [0, -1, 0]
])

sharpened_image = cv2.filter2D(image, -1, sharpen_kernel)

smoothed_image = cv2.GaussianBlur(image, (9,9), 0)

combined_image = cv2.addWeighted(sharpened_image, 0.5, smoothed_image, 0.5, 0)

cv2.imshow("Original Image", image)
cv2.imshow("Sharpened Image", sharpened_image)
cv2.imshow("Smoothed Image", smoothed_image)
cv2.imshow("Combined Image", combined_image)

cv2.imwrite("sharpened_image.jpg", sharpened_image)
cv2.imwrite("smoothed_image.jpg", smoothed_image)
cv2.imwrite("combined_image.jpg", combined_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
